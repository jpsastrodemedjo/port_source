```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

plt.plot(range(0,10),'ro--')
plt.xlabel('abscissa/units')
plt.ylabel('ordinate/units')
plt.grid(True)
```

```python
markers = ['-','+','o','.',',','s','d','x','v','^','<','>','p']

Range = np.random.RandomState(0)
for i in markers:
    plt.plot(Range.rand(2), Range.rand(2), i,
             label="marker='{0}'".format(i))

plt.legend(loc='best',frameon=False);
```
